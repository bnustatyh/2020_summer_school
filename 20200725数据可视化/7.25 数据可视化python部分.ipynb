### 单组柱状图

```python
#控制图片大小以及清晰度 
fig = plt.figure(figsize=(8,4), dpi=72)#150 演示

# 构建数据
x_data = ['第一类','第二类', '第三类','第四类','第五类']
y_data =  np.random.random(5)

# 绘图
plt.bar(x=x_data, height=y_data, color='#02a4de', alpha=0.7)#颜色，透明度  演示alpha
plt.tick_params(labelsize=12)#标签字体大小 演示

# 在柱状图上显示具体数值, ha参数控制水平对齐方式, va控制垂直对齐方式，百分数的表示方法
for x, y in enumerate(y_data):
    plt.text(x,y + 0.005, str('{:.2f}'.format(y *100)) +'%',ha='center',size = 12) #位置，高度，内容，居中

#演示y,ha

#设置y轴范围
plt.ylim(0,1.2)#演示
    
# 设置标题
plt.title('图1 单变量柱状图 ',size = 15)

# 设置坐标轴标题
plt.xlabel(" ",size = 12)
plt.ylabel(" ",size = 12)

#画图
plt.show()
```

### 双组柱状图

```python
#控制图片大小以及清晰度 
fig = plt.figure(figsize=(8,5), dpi=72)

# 构建数据
x = np.arange(5)#一共5组
tick_label = ["A", "B", "C", "D", "E"]#每一组的名称
y = np.random.randint(1,10,5)
y1 = np.random.randint(1,10,5)#每一组数据

#设置柱体宽度
bar_width = 0.35

#绘图
plt.bar(x, y, bar_width, align="center", color="gold", label="组A", alpha=1)
plt.bar(x+bar_width, y1, bar_width, color="grey", align="center", label="组B", alpha=0.7)

#设置y轴范围
plt.ylim(0,11)
   
# 设置标题
plt.title('图2 双变量柱状图 ',size = 15)  

#坐标轴名称
plt.xlabel("类型")
plt.ylabel("数量")

#把X轴的名称换成ABCDE
plt.xticks(x+bar_width/2, tick_label)

#设置图例位置
plt.legend(loc="upper right",borderaxespad=0.2,fontsize = 12)
#upper-center-lower,left-center-right


plt.show()
```

### 堆积柱状图

如果说双柱状图是把X轴的位置换了，那么堆积图就是把Y轴的位置换了，用到的就是Bottom这个参数

```python
#控制图片大小以及清晰度 
fig = plt.figure(figsize=(6,4), dpi=72)

#数据
x = ['第一类','第二类', '第三类','第四类']
y_data1 = [1,3,5,7]
y_data2 = [6,8,2,4]

#z_data = list(np.sum([y_data1,y_data2], axis = 0))

plt.bar(x, y_data1, label='Y1', color='#02a4de',alpha = 1 )  #透明度设置不同的颜色
plt.bar(x, y_data2, label='Y2',bottom=y_data1, color='#02a4de',alpha = 0.7 )
plt.tick_params(labelsize=12)

#数据标签
for x, a in enumerate(y_data1):
    plt.text(x,0.1 , a,ha='center',size = 12) #位置，高度，内容，居中
for x, b in enumerate(y_data2):
    plt.text(x,7,b,ha='center',va = 'bottom',size = 12)

#y轴范围
plt.ylim(0,15)

# 设置标题
plt.title('图3 堆积柱状图 ',size = 15) 

# 为两条坐标轴设置名称
plt.xlabel("类型 ",size = 15)
plt.ylabel("数量 ",size = 15)

# 显示图例
legend = plt.legend(loc='best',borderaxespad=0.2,fontsize = 10) 


plt.show()
```

## 组合：柱状图和折线图组合—双轴的情况

```python
#控制图片大小以及清晰度 
fig = plt.figure(figsize=(10,5), dpi=72)


#--------------设置数据-----------------

#柱状图
a=np.random.randint(25,50,15)

#折线图
b=np.random.random(15)
for i in range(14):
    x = float(a[i+1])
    y = float(a[i])
    b[i+1] = (x-y)/x
b[0] = 0
b = b*100

#标签
x=[i for i in range(15)]
#---------------以上为数据---------------


#设置柱状图的坐标轴
ax1 = fig.add_subplot(111)  #a three-digit number
#111”表示“1×1网格，第一子图”，“234”表示“2×3网格，第四子图

plt.bar(x,a,alpha=0.7,color='#029bbb',label=u'收入（元）') 
ax1.set_ylabel('收入（元）',size =13)
ax1.set_xlabel('年份',size =13)
ax1.set_ylim([0, 60])  #设置y轴取值范围
plt.legend(loc="upper left")


#设置折线图的坐标轴——百分比形式的坐标轴
ax2 = ax1.twinx() #对ax1调用twinx()方法，生成如同镜面效果后的ax2
ax2.plot(x, b,'or-',color='#29527e',label=u'增长率(%)')
fmt='%.f%%'
yticks = mtick.FormatStrFormatter(fmt)
ax2.yaxis.set_major_formatter(yticks)
ax2.set_ylim([-100, 100])
plt.legend(loc="upper left", bbox_to_anchor=(0,0.93))


#刻度字体大小
plt.tick_params(labelsize=15) 


# 设置标题
plt.title('组合图 ',size = 15) 

plt.show()
```

### 饼图

```python
#控制图片大小以及清晰度 
fig = plt.figure(figsize=(4,4), dpi=72)#将画布设定为正方形，则绘制的饼图是正圆

#设置数据
label=['A市','B市','C市']#定义饼图的标签，标签是列表
explode=[0.01,0.01,0.01]#设定各项距离圆心n个半径  #演示尝试突出显示
values = np.random.randint(1,10,3)
colors=['#B2E3F5','#4DBFE7','#02A4DE']

#绘制饼图
plt.pie(values,explode=explode,colors=colors,autopct='%1.2f%%')# shadow=True增加阴影

#绘制标题labels=label,
plt.title('图6 饼图')

#绘制图例
plt.legend(label,loc="upper right", bbox_to_anchor=(1.25,0.4))

plt.show()
```

### 环形图

```python
#控制图片大小以及清晰度 
fig = plt.figure(figsize=(4,4), dpi=72)#将画布设定为正方形，则绘制的饼图是正圆

#设置数据
label = ['A市','B市','C市']
values = np.random.randint(1,10,3)
colors=['#B2E3F5','#4DBFE7','#02A4DE']

#绘图
wedges, texts, autotexts = plt.pie(values,
    autopct = '%3.1f%%',
    radius = 1,                                   #圆的半径
    pctdistance = 0.75,                           #数据标签的半径
    colors = colors,                              #每一部分的颜色
    startangle = 90,                              #起始角度
    textprops = {'color': 'black'},               #数据标签的颜色
    wedgeprops = {'width': 0.5, 'edgecolor': 'w'} #环的宽度，以及边框颜色
)

# 图例
plt.legend(wedges,
           label,
           fontsize = 12,
           title = '地区',
           loc = 'center right',
           bbox_to_anchor = (1.25,0.4))

# 设置文本样式
plt.setp(autotexts, size=15, weight='bold')

# 标题
plt.title('环形图', fontsize=20)

plt.show()
```

## 极坐标绘图

* 在平面内取一个定点O，叫极点，引一条射线Ox，叫做极轴，再选定一个长度单位和角度的正方向（通常取逆时针方向）。对于平面内任何一点M，用ρ表示线段OM的长度（有时也用r表示），θ表示从Ox到OM的角度，ρ叫做点M的极径，θ叫做点M的极角，有序数对 (ρ,θ)就叫点M的极坐标，这样建立的坐标系叫做极坐标系。[百度百科]

* 主要函数 plt.polar

### 简单尝试

```python
n = 10
#theta指每个标记所在射线与极径的夹角，下面表示均分角度

theta = np.linspace(0.0, 2*np.pi, n, endpoint=False)

#r表示点距离圆心的距离，np.random.rand(barSlices)表示返回返回服从“0-1”均匀分布的随机样本值

r = 30*np.random.rand(n)

#polar表示绘制极坐标图，颜色，线宽，标志点样式

plt.polar(theta, r, color="#5596CB", linewidth=2, marker="*", mfc="w", ms=12)#ms标记的大小
```

* theta极角确定有几个点，点的角度，r极径确定长度

### 心形线

* 参数方程的公式
$$x=a(2cos(t)-cos(2t))$$
$$y=a(2sin(t)-sin(2t))$$

* 极坐标公式
$$r = a(1-\sin(\theta))$$

* 用上面的方法

```python
#theta指每个标记所在射线与极径的夹角，下面表示均分角度

theta = np.linspace(0, 2 * np.pi, 1024)

#r表示点距离圆心的距离，np.random.rand(barSlices)表示返回返回服从“0-1”均匀分布的随机样本值
a = 1
r = a*(1. - np.sin(theta))

#polar表示绘制极坐标图，颜色，线宽，标志点样式

plt.polar(theta, r, color="#CC5656", linewidth=1, marker=" ", mfc="w")
```

* 另一种方法

```python
t = np.linspace(0, 2 * np.pi, 1024)
plt.axes(polar = True)
plt.plot(t, 1. - np.sin(t),color="#CC5656")
```
